```python
# Loading the dataset
data = pd.read_csv('./data/TheoryAssignment8-scinameUUIDs-csv.csv', encoding = "ISO-8859-1")

data.head(10)
```

```python
def define_season(row):
    if 'united states' in row['country'].lower() and (row['mo'] == 12 and row['mo'] < 3): 
        return 'winter'
    elif 'united states' in row['country'].lower() and (row['mo'] >= 3 and row['mo'] < 6):
        return 'spring'
    elif 'united states' in row['country'].lower() and (row['mo'] >= 6 and row['mo'] < 9):
        return 'summer'
    elif 'united states' in row['country'].lower() and (row['mo'] >= 9 and row['mo'] < 12):
        return 'fall'
    elif 'australia' in row['country'].lower() and (row['mo'] >= 6 and row['mo'] < 9):
        return 'winter'
    elif 'australia' in row['country'].lower() and (row['mo'] >= 9 and row['mo'] < 12):
        return 'spring'
    elif 'australia' in row['country'].lower() and (row['mo'] == 12 and row['mo'] < 3):
        return 'summer'
    elif 'australia' in row['country'].lower() and (row['mo'] >= 3 and row['mo'] < 6):
        return 'fall'
    else:
        return 'summer'
        
data['seasons'] = data.apply(lambda row: define_season(row), axis=1)
```
