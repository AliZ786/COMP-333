```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
data = pd.read_csv('./data/TheoryAssignment8-scinameUUIDs-csv.csv', encoding = "ISO-8859-1")

data.head(10)
```

```python
data['country']
```

```python
def define_season(row):
    if isinstance(row['country'], str) and 'united states' in row['country'].lower() and (row['mo'] == 12 and row['mo'] < 3): 
        return 'winter'
    elif isinstance(row['country'], str) and 'united states' in row['country'].lower() and (row['mo'] >= 3 and row['mo'] < 6):
        return 'spring'
    elif isinstance(row['country'], str) and 'united states' in row['country'].lower() and (row['mo'] >= 6 and row['mo'] < 9):
        return 'summer'
    elif isinstance(row['country'], str) and 'united states' in row['country'].lower() and (row['mo'] >= 9 and row['mo'] < 12):
        return 'fall'
    elif isinstance(row['country'], str) and 'australia' in row['country'].lower() and (row['mo'] >= 6 and row['mo'] < 9):
        return 'winter'
    elif isinstance(row['country'], str) and 'australia' in row['country'].lower() and (row['mo'] >= 9 and row['mo'] < 12):
        return 'spring'
    elif isinstance(row['country'], str) and 'australia' in row['country'].lower() and (row['mo'] == 12 and row['mo'] < 3):
        return 'summer'
    elif isinstance(row['country'], str) and 'australia' in row['country'].lower() and (row['mo'] >= 3 and row['mo'] < 6):
        return 'fall'
    else:
        return 'summer'
```

```python
data['seasons'] = data.apply(lambda row: define_season(row), axis=1)
```

```python
data
```

```python
# Group the data by season and species, and count the number of occurrences
species_count = data.groupby(['seasons', 'species']).size().reset_index(name='species_count_per_season')

# Merge the counts back into the original DataFrame
data = pd.merge(data, species_count, on=['seasons', 'species'], how='left')
```

```python
data
```

```python
# Create a dictionary to map seasons to numeric values
season_map = {'winter': 1, 'spring': 2, 'summer': 3, 'fall': 4}

# Apply the mapping to the season column
species_count['season_num'] = species_count['seasons'].map(season_map)

# Set the figure size
plt.figure(figsize=(20, 8), dpi=100)

# Create a linear regression plot
sns.lmplot(x='species_count_per_season', y='season_num', data=species_count)

# Set plot title and axis labels
plt.title("Species count per season linear regression", fontdict={'size': 14})
plt.xlabel("Species count per season", fontdict={'size': 14})
plt.ylabel("Season", fontdict={'size': 14})

# Show the plot
plt.show()
```
