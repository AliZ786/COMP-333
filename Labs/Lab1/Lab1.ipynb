```python
# Function that converts a farenheit temperature to a celsius one

def fahr_to_celsius(temp):
    return ((temp - 32) * (5/9))
```

```python
# Checks if 32 degrees farenheit is indeed 0 degrees celsius

fahr_to_celsius(32)
```

```python
# Prints the freezing and boiling points of water 

print('The freezing point of water is:', fahr_to_celsius(32), 'degrees Celsius')
print('The boiling point of water is:', fahr_to_celsius(212), 'degrees Celsius')
```

```python
# Function that returns the average, max and min values from a given dataset

def visualize(filename):

    # loads data 
    data = np.loadtxt(fname=filename, delimiter=',')

    # creates the dimension of the figures
    fig = plt.figure(figsize=(10.0, 3.0))

    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    # determines labels for average 
    axes1.set_ylabel('Average')
    axes1.plot(np.mean(data, axis=0))

    # determines labels for max
    axes2.set_ylabel('Max')
    axes2.plot(np.max(data, axis=0))

    # determines labels for min
    axes3.set_ylabel('Min')
    axes3.plot(np.min(data, axis=0))

    # packages everything together and then prints out the plots
    fig.tight_layout()
    plt.show()
```

```python
# function that checks for anomalies in the min and max plots 

def detect_problems(filename):

    # loads data 
    data = np.loadtxt(fname=filename, delimiter=',')

    # checks if max is suspicious
    if np.max(data, axis=0)[0] == 0 and np.max(data, axis=0)[20] == 20:
        print('Suspicious looking maxima!')

    # checks if min is suspicious    
    elif np.sum(np.min(data, axis=0)) == 0:
        print('Minima add up to zero!')
    
    # else everything is ok 
    else:
        print('Seems OK!')
```

```python
# iterates through the given filename 
# please note that for the TA's remove the ./data in the path since this is my local path only
filenames = sorted(glob.glob('./data/inflammation*.csv'))

for filename in filenames[:3]:
    print(filename)
    visualize(filename)
    detect_problems(filename)
```
